# Import some liberaries

```python
%matplotlib notebook
import netCDF4
from netCDF4 import Dataset
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import seaborn as sns
```

# Read land, coast, and sea data of TRMM with respect to MJO phases

```python
f1 = Dataset('WS12_er_negative_land.nc', mode='r')
f2 = Dataset('WS12_er_positive_land.nc', mode='r')
f3 = Dataset('WS12_kelvin_negative_land.nc', mode='r')
f4 = Dataset('WS12_kelvin_positive_land.nc', mode='r')
f5 = Dataset('WS12_mjo_negative_land.nc', mode='r')
f6 = Dataset('WS12_mjo_positive_land.nc', mode='r')

data1 = [f1.variables['IRregime'],f2.variables['IRregime'],f3.variables['IRregime'],f4.variables['IRregime'], \
       f5.variables['IRregime'],f6.variables['IRregime']]

f9 =  Dataset('WS12_er_negative_coast.nc', mode='r')
f10 = Dataset('WS12_er_positive_coast.nc', mode='r')
f11 = Dataset('WS12_kelvin_negative_coast.nc', mode='r')
f12 = Dataset('WS12_kelvin_positive_coast.nc', mode='r')
f13 = Dataset('WS12_mjo_negative_coast.nc', mode='r')
f14 = Dataset('WS12_mjo_positive_coast.nc', mode='r')

data2 = [f9.variables['IRregime'],f10.variables['IRregime'],f11.variables['IRregime'],f12.variables['IRregime'], \
       f13.variables['IRregime'],f14.variables['IRregime']]

f17 = Dataset('WS12_er_negative_sea.nc', mode='r')
f18 = Dataset('WS12_er_positive_sea.nc', mode='r')
f19 = Dataset('WS12_kelvin_negative_sea.nc', mode='r')
f20 = Dataset('WS12_kelvin_positive_sea.nc', mode='r')
f21 = Dataset('WS12_mjo_negative_sea.nc', mode='r')
f22 = Dataset('WS12_mjo_positive_sea.nc', mode='r')

data3 = [f17.variables['IRregime'],f18.variables['IRregime'],f19.variables['IRregime'],f20.variables['IRregime'], \
       f21.variables['IRregime'],f22.variables['IRregime']]
```

# Plot the data using panel plots of box plot

```python
# define function to plot a figure
fig = plt.figure(figsize=(7,14))

# Boxplot of TMPA-MJO over Land
plt.subplot(3, 1, 1)
plt.boxplot(data1, showfliers=False, showmeans=True)
plt.title('ISCCP IR-WS12 and Equatorial Waves over the Maritime Continent')
plt.ylabel('Frequency (%)')
plt.xticks(x,"")
plt.text(1.0, 1.1, 'a) Land', fontsize=15)
plt.ylim([-0.1, 1.2])

# Boxplot of TMPA-MJO over Coast
plt.subplot(3, 1, 2)
plt.boxplot(data2, showfliers=False, showmeans=True)
plt.ylabel('Frequency (%)')
plt.xticks(x,"")
plt.text(1.0, 1.1, 'b) Coast', fontsize=15)
plt.ylim([-0.1, 1.2])

# Boxplot of TMPA-MJO over Sea
plt.subplot(3, 1, 3)
plt.boxplot(data3, showfliers=False, showmeans=True)
plt.ylabel('Frequency (%)')
x=np.array([1,2,3,4,5,6])
my_xticks = ['ER +ve','ER -ve','Kelvin +ve','Kelvin -ve','MJO +ve','MJO -ve']
plt.xticks(x,my_xticks)
plt.text(1.0, 1.1, 'c) Sea', fontsize=15)
plt.ylim([-0.1, 1.2])
plt.xlabel('Equatorial Waves Standard Anomaly')

# Customize the plot and save 

plt.subplots_adjust(hspace=0.05)
plt.show()
plt.savefig('Diurnal_IRWS12_waves_boxplot.png')
```
