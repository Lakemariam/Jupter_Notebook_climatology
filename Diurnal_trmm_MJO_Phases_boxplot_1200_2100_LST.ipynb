# Plot the data using panel plots of box plot

```python
# set the panel plot size and orientation
# write the title of respective panel plots

fig, ax = plt.subplots(3,2, sharex=True, sharey=True, figsize=(12, 12))
font = FontProperties()
font.set_size('xx-large')

fig.text(0.5, 0.04, 'MJO Phases', ha='center', fontproperties=font)
fig.text(0.04, 0.5, 'Precipitation (mm/hr)', va='center', rotation='vertical', fontproperties=font)

# Assign position of each plots to be assigned
ax[0,0].boxplot(data1, showfliers=False)
ax[0,0].text(0.7, 1.0, 'a) Land', fontsize=15)
ax[0,0].set_title('TMPA and MJO Phases at 1200 LST', fontproperties=font)
ax[0,0].axis([1,8,0,1.1])

ax[1,0].boxplot(data2, showfliers=False)
ax[1,0].text(0.7, 1.0, 'b) Coast', fontsize=15)

ax[2,0].boxplot(data3, showfliers=False)
ax[2,0].text(0.7, 1.0, 'c) Sea', fontsize=15)

# ssecond column
# Assign position of each plots to be assigned
ax[0,1].boxplot(data4, showfliers=False)
ax[0,1].text(0.7, 1.0, 'a) Land', fontsize=15)
ax[0,1].set_title('TMPA and MJO Phases at 2100 LST', fontproperties=font)

ax[1,1].boxplot(data5, showfliers=False)
ax[1,1].text(0.7, 1.0, 'b) Coast', fontsize=15)

ax[2,1].boxplot(data6, showfliers=False)
ax[2,1].text(0.7, 1.0, 'c) Sea', fontsize=15)

x=np.array([1,2,3,4,5,6,7,8])
my_xticks = ['Ph1','Ph2','Ph3','Ph4','Ph5','Ph6','Ph7','Ph8']
plt.xticks(x,my_xticks)

# Customize panel plot

plt.subplots_adjust(hspace=0.05, wspace=0.05)
plt.show()
plt.savefig('Diurnal_trmm_MJO_Boxplot_12_21LST.png')

```
