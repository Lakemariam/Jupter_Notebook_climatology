Import some liberaries

```python
%matplotlib notebook
import netCDF4
from netCDF4 import Dataset
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import seaborn as sns
```

Read land, coast, and sea data of TRMM with respect to MJO phases

```python
f1 = Dataset('trmm_fix_djfph1_land.nc', mode='r')
f2 = Dataset('trmm_fix_djfph2_land.nc', mode='r')
f3 = Dataset('trmm_fix_djfph3_land.nc', mode='r')
f4 = Dataset('trmm_fix_djfph4_land.nc', mode='r')
f5 = Dataset('trmm_fix_djfph5_land.nc', mode='r')
f6 = Dataset('trmm_fix_djfph6_land.nc', mode='r')
f7 = Dataset('trmm_fix_djfph7_land.nc', mode='r')
f8 = Dataset('trmm_fix_djfph8_land.nc', mode='r')

data1 = [f1.variables['precip'],f2.variables['precip'],f3.variables['precip'],f4.variables['precip'], \
       f5.variables['precip'],f6.variables['precip'],f7.variables['precip'],f8.variables['precip']]

f9 =  Dataset('trmm_fix_djfph1_coast.nc', mode='r')
f10 = Dataset('trmm_fix_djfph2_coast.nc', mode='r')
f11 = Dataset('trmm_fix_djfph3_coast.nc', mode='r')
f12 = Dataset('trmm_fix_djfph4_coast.nc', mode='r')
f13 = Dataset('trmm_fix_djfph5_coast.nc', mode='r')
f14 = Dataset('trmm_fix_djfph6_coast.nc', mode='r')
f15 = Dataset('trmm_fix_djfph7_coast.nc', mode='r')
f16 = Dataset('trmm_fix_djfph8_coast.nc', mode='r')

data2 = [f9.variables['precip'],f10.variables['precip'],f11.variables['precip'],f12.variables['precip'], \
       f13.variables['precip'],f14.variables['precip'],f15.variables['precip'],f16.variables['precip']]

f17 = Dataset('trmm_fix_djfph1_sea.nc', mode='r')
f18 = Dataset('trmm_fix_djfph2_sea.nc', mode='r')
f19 = Dataset('trmm_fix_djfph3_sea.nc', mode='r')
f20 = Dataset('trmm_fix_djfph4_sea.nc', mode='r')
f21 = Dataset('trmm_fix_djfph5_sea.nc', mode='r')
f22 = Dataset('trmm_fix_djfph6_sea.nc', mode='r')
f23 = Dataset('trmm_fix_djfph7_sea.nc', mode='r')
f24 = Dataset('trmm_fix_djfph8_sea.nc', mode='r')

data3 = [f17.variables['precip'],f18.variables['precip'],f19.variables['precip'],f20.variables['precip'], \
       f21.variables['precip'],f22.variables['precip'],f23.variables['precip'],f24.variables['precip']]
```

Plot the data using panel plots of box plot

```python
# define function to plot a figure
fig = plt.figure(figsize=(7,14))

# Boxplot of TMPA-MJO over Land
plt.subplot(3, 1, 1)
plt.boxplot(data1, showfliers=False)
plt.title('TMPA Rainfall and MJO Phases over the Maritime Continent')
plt.ylabel('Precipitation (mm/hr)')
plt.xticks(x,"")
plt.text(1.0, 1.0, 'a) Land', fontsize=15)
plt.ylim([-0.1, 1.1])

# Boxplot of TMPA-MJO over Coast
plt.subplot(3, 1, 2)
plt.boxplot(data2, showfliers=False)
plt.ylabel('Precipitation (mm/hr)')
plt.xticks(x,"")
plt.text(0.8, 0.8, 'b) Coast', fontsize=15)
plt.ylim([-0.1, 0.9])

# Boxplot of TMPA-MJO over Sea
plt.subplot(3, 1, 3)
plt.boxplot(data3, showfliers=False)
plt.ylabel('Precipitation (mm/hr)')
x=np.array([1,2,3,4,5,6,7,8])
my_xticks = ['Phase1','Phase2','Phase3','Phase4','Phase5','Phase6','Phase7','Phase8']
plt.xticks(x,my_xticks)
plt.text(0.75, 0.75, 'c) Sea', fontsize=15)
plt.ylim([-0.1, 0.9])
plt.xlabel('MJO Phases')

# Customize the plot and save 

plt.subplots_adjust(hspace=0.05)
plt.show()
plt.savefig('Diurnal_trmm_MJO_Percentile.png')
```
