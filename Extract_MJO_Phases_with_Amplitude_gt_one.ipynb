```python
#df = get_RMM()
import os
os.getcwd()

def get_RMM(url='http://cawcr.gov.au/staff/mwheeler/maproom/RMM/',fname='RMM1RMM2.74toRealtime.txt',start=1974, end=2018):
    """
    # Function to load the MJO index(es) from Wheeler and Hendon (RMM1 and RMM2) from the URL
    
    """
    data = pd.read_table('%s%s' % (url, fname), sep='\s*', skiprows=2, header=None, \
                           names = ['year','month','day','RMM1', 'RMM2', 'MJO phase', 'amplitude', 'origin'])
    dates = [datetime(y,m,d) for y,m,d in zip(data.year, data.month, data.day)]
    data.index = dates
    data = data.truncate(before="%s-1-1" % ( start ), after= "%s-12-31" % ( end ))
    return data
```

## Function to import text data and process composite

```python
from datetime import datetime
def processes_txt():
    """
    This function imports text data to read and coverts it
    into dataframe for ease of manimpulation
    """
    data = pd.read_table('rmm.74toRealtime.txt', sep='\s*', skiprows=2, header=None, \
                           names = ['year','month','day','RMM1', 'RMM2', 'MJO phase', 'amplitude', 'origin'], engine='python')
    dates = [datetime(y,m,d) for y,m,d in zip(data.year, data.month, data.day)]
    data.index = dates
    return data
data = processes_txt()
data.head()
```

```python
def slice_filter(data):
    """
    This function will take the dataframe output from the above function
    and slice the date into the desired period of study (e.g. 1984-2009) and
    filter the data according to the amplitude variable and next will classies 
    the data according to the phase variable
    """
    df_sliced = data.loc['19830101':'20081231']
    df_amp = df_sliced[df_sliced['amplitude']>=1]
    df_ph1 = df_amp[df_amp['MJO phase']==1]
    df_ph2 = df_amp[df_amp['MJO phase']==2]
    df_ph3 = df_amp[df_amp['MJO phase']==3]
    df_ph4 = df_amp[df_amp['MJO phase']==4]
    df_ph5 = df_amp[df_amp['MJO phase']==5]
    df_ph6 = df_amp[df_amp['MJO phase']==6]
    df_ph7 = df_amp[df_amp['MJO phase']==7]
    df_ph8 = df_amp[df_amp['MJO phase']==8]
    return df_ph1,df_ph2,df_ph3,df_ph4,df_ph5,df_ph6,df_ph7,df_ph8
    
df_ph1,df_ph2,df_ph3,df_ph4,df_ph5,df_ph6,df_ph7,df_ph8 =   slice_filter(data)  
```

```python
def seasonal_compisite(df,jja =[6,7,8],son = [9,10,11],djf=[12,1,2],mam=[3,4,5]):
    """
    This function will take the phase data that obtained in the above function
    and outputs date indexes used  for composite analysis later
    """
    df = df.reset_index()
    df_jja = df[(df['month']==jja[0]) | (df['month']==jja[1])|(df['month']==jja[2])]
    df_son = df[(df['month']==son[0]) | (df['month']==son[1])|(df['month']==son[2])]
    df_mam = df[(df['month']==mam[0]) | (df['month']==mam[1])|(df['month']==mam[2])]
    df_djf = df[(df['month']==djf[0]) | (df['month']==djf[1])|(df['month']==jja[2])]
    #df_jja['index'].to_csv(filename,sep=',')
    return df_jja,df_son,df_mam,df_djf

df_jja,df_son,df_mam,df_djf = seasonal_compisite(df_ph8,jja =[6,7,8],son = [9,10,11],djf=[12,1,2],mam=[3,4,5])

df_jja['index'].to_csv('jja_irws_ph8.csv',sep=',')
df_mam['index'].to_csv('mam_irws_ph8.csv',sep=',')
df_son['index'].to_csv('son_irws_ph8.csv',sep=',')
df_djf['index'].to_csv('djf_irws_ph8.csv',sep=',')
df_djf.shape
```
