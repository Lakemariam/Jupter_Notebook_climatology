```python
%matplotlib notebook
import netCDF4
from netCDF4 import Dataset
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
f1 = Dataset('trmm_fix_djfph1_land.nc', mode='r')
f2 = Dataset('trmm_fix_djfph2_land.nc', mode='r')
f3 = Dataset('trmm_fix_djfph3_land.nc', mode='r')
f4 = Dataset('trmm_fix_djfph4_land.nc', mode='r')
f5 = Dataset('trmm_fix_djfph5_land.nc', mode='r')
f6 = Dataset('trmm_fix_djfph6_land.nc', mode='r')
f7 = Dataset('trmm_fix_djfph7_land.nc', mode='r')
f8 = Dataset('trmm_fix_djfph8_land.nc', mode='r')

f9 = Dataset('trmm_fix_djfph1_coast.nc', mode='r')
f10 = Dataset('trmm_fix_djfph2_coast.nc', mode='r')
f11 = Dataset('trmm_fix_djfph3_coast.nc', mode='r')
f12 = Dataset('trmm_fix_djfph4_coast.nc', mode='r')
f13 = Dataset('trmm_fix_djfph5_coast.nc', mode='r')
f14 = Dataset('trmm_fix_djfph6_coast.nc', mode='r')
f15 = Dataset('trmm_fix_djfph7_coast.nc', mode='r')
f16 = Dataset('trmm_fix_djfph8_coast.nc', mode='r')

f17 = Dataset('trmm_fix_djfph1_sea.nc', mode='r')
f18 = Dataset('trmm_fix_djfph2_sea.nc', mode='r')
f19 = Dataset('trmm_fix_djfph3_sea.nc', mode='r')
f20 = Dataset('trmm_fix_djfph4_sea.nc', mode='r')
f21 = Dataset('trmm_fix_djfph5_sea.nc', mode='r')
f22 = Dataset('trmm_fix_djfph6_sea.nc', mode='r')
f23 = Dataset('trmm_fix_djfph7_sea.nc', mode='r')
f24 = Dataset('trmm_fix_djfph8_sea.nc', mode='r')
```

```python
fig = plt.figure(figsize=(8,4))
x=np.array([0,1,2,3,4,5,6,7])
plt.plot(x, [np.average(f1.variables['precip']), np.average(f2.variables['precip']), np.average(f3.variables['precip']), \
             np.average(f4.variables['precip']), np.average(f5.variables['precip']), np.average(f6.variables['precip']), \
             np.average(f7.variables['precip']), np.average(f8.variables['precip'])], color="black", label="Land")

plt.plot(x, [np.average(f9.variables['precip']), np.average(f10.variables['precip']), np.average(f11.variables['precip']), \
             np.average(f12.variables['precip']), np.average(f13.variables['precip']), np.average(f14.variables['precip']), \
             np.average(f15.variables['precip']), np.average(f16.variables['precip'])], color="red", label="Coast")

plt.plot(x, [np.average(f17.variables['precip']), np.average(f18.variables['precip']), np.average(f19.variables['precip']), \
             np.average(f20.variables['precip']), np.average(f21.variables['precip']), np.average(f22.variables['precip']), \
             np.average(f23.variables['precip']), np.average(f24.variables['precip'])], color="blue", label="Sea")
             
my_xticks = ['Phase1','Phase2','Phase3','Phase4','Phase5','Phase6','Phase7','Phase8']
plt.xticks(x,my_xticks)
plt.xlabel('MJO Phases')
plt.ylabel('Precipitation (mm/hr)')
plt.title('Diurnal Mean TMPA Rainfall and MJO Phases over the MC')
plt.legend()
plt.show()
plt.savefig('Diurnal_average_trmm_MJO_phases.png')
```
